#Problem Statement 2: Write a Python program to perform the following tasks
1.Create a pandasseries from the below dictionary where indices are subjects:
{
'English':{'Sam':60,'Jackson':74,'Ahree':85}, 
'History':{'Gloria':83,'Sam':65,'Isla':78,'Aron':72,'Gray':61},
'Geography':{'Jackson':92,'Gloria':95,'Isla':82,'Aron':75,'Ahree':76},'Mathematics':{'Sam':99,'Gloria':74,'Jackson':89,'Ahree':85,'Gray':95},'Science':{'Sam':89,'Aron':82,'Gray':78,'Isla':93,'Ahree':87}
}
2.Convert the created series into DataFrame and replace the null values with zeroes.
3.Transpose the DataFrame and create a new column 'Average' and fill the values in it by calculating the average of all subjects

```python
import pandas as pd
import numpy as np
```

```python
#Task1.Create a pandasseries from the below dictionary where indices are subjects:
exam={
    'English':{'Sam':60,'Jackson':74,'Ahree':85},
    'History':{'Gloria':83,'Sam':65,'Isla':78,'Aron':72,'Gray':61},
    'Geography':{'Jackson':92,'Gloria':95,'Isla':82,'Aron':75,'Ahree':76},
    'Mathematics':{'Sam':99,'Gloria':74,'Jackson':89,'Ahree':85,'Gray':95},
    'Science':{'Sam':89,'Aron':82,'Gray':78,'Isla':93,'Ahree':87}
    }
ser=pd.Series(exam)
ser
```

```python
#Task2:Convert the created series into DataFrame
df=ser.apply(pd.Series)
df
```

```python
#Task2.replace the null values with zeroes.
df1=df.fillna(0)
df1
```

```python
#Task3:Transpose the DataFrame 
df_transpose=df1.T
df_transpose
```

```python
#Task3:create a new column 'Average' and fill the values in it by calculating the average of all subjects
df_transpose['Average']=df1.apply(np.mean)
df_transpose
```

#Problem Statement 3: Write a Python program to create a series from 1 to 1000 and select only numbers divisible by 7 and 17

```python
import numpy as np
import pandas as pd
number=pd.Series(np.arange(1,1001))
print(number[(number % 7==0) & (number%17==0)])
```

#Problem Statement 4: Sylphia has a dataset of various cereals sold in the supermarket.
Dataset Description:
cereal.csv - The dataset contains 16 features. Here's a brief description of 3 columns in the
dataset:
• name - Brand name of the cereals
• MFR - Manufacturer of the brands
• rating - Rating of the cereals
Syliphia wants to visualize the quality of cereals and determine which manufacturer delivers
the best quality.
Tasks to be performed:
1. Import the dataset.
2. Plot ratings of different types of manufacturers.
3. Use xticks range form 0-100.
4. Change the style of the graph to seaborn

```python
import pandas as pd
```

```python
!wget https://www.dropbox.com/s/idnul34dfo5cnke/cereal.csv
```

```python
#Task1:Import the dataset.
market=pd.read_csv('cereal.csv')
market.head()
```

```python
rating = market.groupby('mfr')['rating'].mean()
rating
```

```python
#Task2: Plot ratings of different types of manufacturers.
import matplotlib.pyplot as plt
plt.bar(rating.index, rating.values)
plt.show()
```

```python
#Task3: Use xticks range form 0-100.
import matplotlib.pyplot as plt
plt.barh(rating.index, rating.values)
plt.xlim(0,100)
plt.show()
```

```python
#Task4: Change the style of the graph to seaborn
plt.style.use('seaborn')
plt.barh(rating.index, rating.values)
plt.xlim(0,100)
plt.show()
```
